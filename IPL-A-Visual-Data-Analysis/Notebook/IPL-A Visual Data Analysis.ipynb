## Asking and Answering Questions

**As of now we have seen various plots and visual reprentations for the whole data frame. to further analyse the set , it becomes important to work with a small dataframe from the main dataframe i.e. `ipl_df`. ### North- South Rivalry

Being big fans of the game , we are aware of the North south rivalries of the teams . Here is a try to answer a few of the fundamental and basic questions to learn about the data. First subdomain dataframe selected is **KXIP v/s CSK.**

```python
KXIP_CSK=ipl_df[np.logical_or(np.logical_and(ipl_df['team1']=='KXIP',ipl_df['team2']=='CSK'),np.logical_and(ipl_df['team2']=='KXIP',ipl_df['team1']=='CSK'))]
```

```python
KXIP_CSK.sample(5)  #Sample shows a few of the data points in the wole dataframe. Here, number shown is 5. 
```

```python
KXIP_CSK.shape
```

### Q.1. How has winning a toss helped winning the game ?

```python
w=0
l=0
for win in range(0,21):
    if KXIP_CSK.toss_winner[win]==KXIP_CSK.winner[win]:
        w+=1;
    else:
        l+=1
print('Match winner with toss win=',w,', Total matches=',w+l,',  Win Percentage=',w*100/(w+l))
```

**The above question is answered using a nested `for` loop with the `if-else` statement to aid it .** The answer tells us that winning a toss has helped the teams register victory 61.9% of times i.e. 13 wins of 21.

The decisions taken by each team captain upon winning a toss depends on various factors. Hence, it is important to see how each ground makes the captain choose between the options of **BATTING** and **BOWLING**.

### Q.6.How has homeground / home_city affected their performances?

Each team plays half of its matches at home and other half away from home. Therefore, it's neccessary that the team performs better at home than the rest of the grounds. **This particular question answers the question of how many matches played on home ground as well as away**. It also tells the percentage of winning at home and away with a nested `for loop` code.

```python
RR_home=KKR_home=0
RR_home_win=RR_away_win=0
KKR_home_win=KKR_away_win=0
for team in range(len(KKR_RR)):
    if KKR_RR.city[team]=='Jaipur' or KKR_RR.city[team]=='Ahmedabad':
        RR_home+=1
        if KKR_RR.winner[team]=='RR':
            RR_home_win+=1;
        else:
            KKR_away_win+=1;
    elif KKR_RR.city[team]=='Kolkata':
        KKR_home+=1
        if KKR_RR.winner[team]=='RR':
            RR_away_win+=1;
        else:
            KKR_home_win+=1;

print(f'RR has won {RR_home_win} matches at home against KKR and win percentage at home is {RR_home_win*100/RR_home}')        
print(f'KKR has won {KKR_home_win} matches at home against RR and win percentage at home is {KKR_home_win*100/KKR_home}')        
print(f'RR away win percentage is {RR_away_win*100/KKR_home} and KKR away win percentage is {KKR_away_win*100/RR_home} with each other.')                
```

KKR has better home ground winning stats in comparison with RR against them.
